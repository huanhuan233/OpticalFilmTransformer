```python
import pandas as pd 
import numpy as np
import os
import pickle as pkl
```

```python
mode = 'dev' # or 'test

data_path = 'optogpt_data/' + mode 

data_names = os.listdir(data_path)

df_list = []

for file in data_names:
    df_ = pd.read_csv(os.path.join(data_path, file))
    df_list.append(df_)
    
df = pd.concat(df_list, ignore_index=True)
```

```python
df
```

```python
import ast

spec_index = ['R_'+str(i)+'nm' for i in range(380, 751, 5)] + ['T_'+str(i)+'nm' for i in range(380, 751, 5)]

spec = df[spec_index].values

structure = list(df['structure'].apply(ast.literal_eval))
```

```python

with open('dataset/Structure_'+mode+'.pkl', 'wb') as files:
    pkl.dump(structure, files)
    
with open('dataset/Spectrum_'+mode+'.pkl', 'wb') as files:
    pkl.dump(spec, files)
```
